```python
# 정치는 중립적이어야하고, 등등........
# 간단하게 명령어하나만치면! 찾고자하는 단어/ 문서에서 가장많은 쌍을이루는 단어를 가져올수잇음.
# 해석은 하기나름.
```

```python
# - opinion 
# - 확률값 개념 들어가지만, 실제 확률을 계산할순없음.
# - 통계 : 가지고잇는 문서에서 몇번나왔는지 빈도수를 계산해서 확률값을 추정함.
# - 두 이산확률변수 간의 상관성을 분석해주는 MI
# => 특정한 값들 중 하나! 를 두고 비교해주는게 PMI"


# 찾고자하는 특정 단어가 긍정이다/부정이다  -> 감성분석.
# 관련이 있나 비교 -> opinion mining
# 특정 단어와 문서간의 관계 여부 -> 관련이 있는 문서 여부 확인가능.


# PMI 분자 : 같이나온 확률
#     분모 : 각각의 확률 곱한거.
# PMI = 0 이면 서로 전혀다른 독립적인 관계.
# -가 되면 서로 관련이 없다.
# +가 되면 같이나온 관련성이 크다.

# 문제점) 수집한 corpus에 따라, 수집한 문서의 부분에 ㄸ라 다른값이 나올수잇어서 
# bias가 발생할수잇음. -무한대.. +무한대나옴.
# -1~+1 사이로 정규화하기도함.
```

```python
# 정보검색...!
# ES 써서하는팀?

# 의미적, 통계적으로 관련있는 애들을 뽑기.
from nltk import collocations
```

```python
corpus
```

```python
# 2개의 단어쌍(window size=2인애)
measure = collocations.BigramAssocMeasures()
```

```python
# 단어랑, 품사랑 태깅된 결과가 나옴.
taggedWords = kkma.pos(corpus)
```

```python
# bigram으로 넣어준거를 여기에 넣어주면, 확률,통계적으로 얼마나 이웃ㅎ서 많이나오는지 보여줌
# measure아니고finder쓸거임. 이미 형태소분석해놔서 words로부터 가져올거임.

finder = collocations.BigramCollocationFinder.from_words(taggedWords)
```

```python
for row in finder.nbest(measure.pmi, 5):
    print(row)

    
# 강제-노역 관련잇음.
```

```python
tags = (t for (w,t) in taggedWords)
```

```python
words = (w for (w,t) in taggedWords)
```

```python
# tag만있고, words만 있도록 분리해놓음.
```

```python
kkma.tagset
```

```python
finder = collocations.BigramCollocationFinder.from_words(tags)

for row in finder.nbest(measure.pmi, 5):
    print(row)
```

```python
# 형식형태소
# 실질혛ㅇ태소 -> 분사 다음엔 뭐가나오더라, 대명사 다음엔 뭐가 나오더라 판단 가능.

# 신조어, 비속어 등은 형태소분석기에 돌리면 일반명사로 추정할가능성이 높음.
# 이런애들의 형태소분석을 제대로 못할수잇음.
# 신조어에 대해서 자동으로 품사를 태깅할 때도 사용가능.

finder = collocations.BigramCollocationFinder.from_words(words)

for row in finder.nbest(measure.pmi, 5):
    print(row)
```

```python
wordList = [k for (k,v) in kkma.pos(corpus)]
wordList
```

```python
# 형식형태소 없애기. 하, ㄴ ... 명사들만 있도록.
wordList = [k for (k,v) in kkma.pos(corpus) if v.startswith('NN')] 
wordList
```

```python
from collections import Counter
```

```python
wordCount = Counter(wordList)
# 명사 List들에다가 몇번 나왔는지 count가 붙어서 나옴.
wordCount = wordCount.most_common(40)
wordCount
# 가장 빈도수가 높은 40만 반환.
```

```python
type(wordCount)
```

```python
import pytagcloud
```

```python
tagList = pytagcloud.make_tags(wordCount, maxsize=40)
tagList
```
