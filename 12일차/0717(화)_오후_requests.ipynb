```python
# 파일을 열기위해서 사용하는 명령어는 open().
# 파일객체랑 제일 처음 인사하는 명령어임.

f = open('moon.txt', 'r', encoding="utf-8")
# 'r' read
# 'a' append
# 'b'
# 'x'
```

```python
type(f)

# 언더바있으니까 내부객체.
# iterator은 for in 뒤에 쓸 수 있음.
```

```python
# 닫기.
f.close()
```

```python
for i in f:
    print(i)
```

```python
files = []
for i in range(10000):
    files.append(open('moon.txt'))
    
# OS 에러. 한꺼번에 열수있는 파일의 개수가 정해져잇음. 그거 이상으로 불렀을 때, 오류남.
# 파일열면 항상 닫아줘야함.
# but, Python은 자동으로 garbage collection되서 메모리상에서 없어짐.
# context Many : 무조건 실행시켜야하므로 try-finally 써줘야함. finally에 닫는처리해주게.
```

```python
# with 쓰면 안닫아줘도됨? 자동으로 닫아줌?? 
f.close()
with open('moon.txt', encoding="utf-8") as f:
    print(f.read())
```

```python
# ㅇㅇ

for i in f:
    print(i)
```
