```python
# super의 인자에 부모를 넣었을 때,

class Penguin(Bird):
    def swim(self):
        print("I can't Swim")
        
    def __init__(self):
        #call super() function
        super(Penguin).__init__()
        print("Penguine is ready")
        
#    def whoisThis(self):
#        print("Penguine")

    def run(self):
        print("Run faster")
Penguin.__mro__
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
```

```python
# super의 인자에 자식을 넣었을 때,

class Penguin(Bird):
    def swim(self):
        print("I can't Swim")
        
    def __init__(self):
        #call super() function
        super(Bird).__init__()
        print("Penguine is ready")
        
#    def whoisThis(self):
#        print("Penguine")

    def run(self):
        print("Run faster")
Penguin.__mro__
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
```

```python
# super의 인자에 부모, 자식을 넣엇을 때,

class Penguin(Bird):
    def swim(self):
        print("I can't Swim")
        
    def __init__(self):
        #call super() function
        super(Penguin, self).__init__()
        print("Penguine is ready")
        
#    def whoisThis(self):
#        print("Penguine")

    def run(self):
        print("Run faster")
Penguin.__mro__
```

```python
peggy = Penguin()
peggy.whoisThis()
peggy.swim()
```

```python
# super의 인자에 부모, 자식을 넣엇을 때,
# 다중상속을 받을때(상속받는 부모클래스가 여러개 있을 때), Python은 왼쪽에 있는 애부터 상속받도록 순서가 있음.

class Penguin(Bird):
    def swim(self):
        print("I can't Swim")
        
    def __init__(self):
        #call super() function
        super(Bird, self).__init__()
        print("Penguine is ready")
        
#    def whoisThis(self):
#        print("Penguine")

    def run(self):
        print("Run faster")
        
peggy = Penguin()
peggy.whoisThis()
peggy.swim()

# super(Bird, self) : 다중상속때문...!! 어떤 부모꺼 상속받겠다고 딱 찍는거임.
# 이걸 안쓰면 왼쪽부터 차례대로 다중 상속받기때문.
```

```python
Penguin.__mro__

# 다중상속일 때, 접근하는받는 메서드의 순서 보여줌.
```

```python
class Computer:

    def __init__(self):
        self.__maxprice = 900
    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

c = Computer()
c.sell()

# change the price
c.__maxprice = 1000
c.sell()

# using setter function
c.setMaxPrice(1000)
c.sell()
```

```python
# Dotyping 기법을 Python에서 제공하기때문에, 상속안받고도 다형성 쓸 수 있음.
# 상속안받아도 충분히 다 할수잇다는 거.
# 상속보다 훨씬 ㅋ는 범위에서 사용가능 .상속 안받아도 쓸수잇음.

class Parrot:

    def fly(self):
        print("Parrot can fly")
    
    def swim(self):
        print("Parrot can't swim")

class Penguin:

    def fly(self):
        print("Penguin can't fly")
    
    def swim(self):
        print("Penguin can swim")
```

```python
# common interface
def flying_test(bird):
    bird.fly()
# doctyping지원안하는 다른언어에선 실행불가...!!


#instantiate objects
blu = Parrot()
peggy = Penguin()

# passing the object
flying_test(blu)
flying_test(peggy)
```
