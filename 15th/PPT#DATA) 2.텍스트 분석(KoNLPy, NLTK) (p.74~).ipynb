# Heap's Law (힙의 법칙)

단어를 쭉 뽑으면 굉장히 많이나오는데 그게 다 필요한게 아니다.
결국 상위 10%가 전체의 정확도를 결정짓는다?
점점 중요도가 떨어지면서 출현빈도가 줄어드는 것을 볼거에요.

```python
from konlpy.corpus import kolaw
from konlpy.tag import Kkma as kkma
from matplotlib import pyplot as plt
```

```python
text = kolaw.open(kolaw.fileids()[0]).read()
```

```python
len(nounList)
```

```python
global_unique = []
```

```python
global_unique_count = []
```

```python
# text를 split하면 용량이 줄어들거임.

for row in text.split("\n"):
    term = t.nouns(row)
    #print(len(term))
    unique = set(term)
    #print(unique)
    global_unique += list(unique)
    # 중복되는 애들 없애주기.
    global_unique = list(set(global_unique))
    #print(global_unique)
    global_unique_count.append(len(global_unique))
    print(len(unique), len(global_unique))
```

```python
plt.plot(global_unique_count)
plt.savefig("tet.png")

# 단어를 쭉 찾다가 어느순간 둔화해서 갈거임. 
# 따라서 어느정도 적당히 나온곳까지만 잘라서 feature로 이용할 수 있다는 것이 Heap's Law.
```

```python
# 읽은 결과를 리스트로 만들고 하나씩 text에 저장할거임.
# 10개의 텍스트 리스트가 생길것임.

text = [kobill.open(i).read for i in kobill.fileids()]
```

```python
# 아까보단 좀 완만해지긴 함.
# 앞으로 새로운문자라 나올 일이 거의 없기때문에 적당한 곳에서 끊으면 됨.
# 텍스트 분석하다보면 어절하나에서 후보가 굉장힘 낳을거임 .이중 몇개를서야하고 노드는몇개를 만들어야하며... 등등
# count가지고 세고 상위 20%. count로 빈도세서 전체 40%까지만 피쳐로 쓴다던가 하는거임.

for row in text:
    term = t.nouns(row)
    unique = set(term)
    global_unique += list(unique)
    global_unique = list(set(global_unique))
    global_unique_count.append(len(global_unique))
    print(len(unique), len(global_unique))
```
