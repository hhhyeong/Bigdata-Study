# 0702(월) 배우고 싶은 것 (정혜영)

저는 새로운 기술에 관심이 많습니다. 따라서 저의 관심분야를 전반적으로 다루는 교육 과정 커리큘럼을 성실하게 이수하며 다양한 가능성을 열어두고, 한 주제가 끝날 때마다 배운것을 활용하여 어떤 프로젝트를 진행할 수 있을지 구체적으로 계획하고 구현해볼 계획입니다. 두달 동안, 막연했던 저의 관심과 혼자 공부하면서 막막했던 부분을 해소하고, 관심분야를 좁힐 수 있을것이라 생각하여 지원했습니다. 청년 빅데이터 아카데미에 오기까지의 과정과 교육을 통해 배우고 싶은 것들은 다음과 같습니다.

먼저, 현업에서 다루는 방대한 양의 데이터를 어떻게하면 효율적으로 다룰 수 있을지에 대한 관심이 많습니다. 4학년 1학기, 소프트웨어 프로젝트 과목에서 K-means clustering 알고리즘을 활용한 데이터 마이닝 기법을 Hadoop의 MapReduce방식으로 구현하는 실습을 통해 빅데이터 처리에 흥미를 갖게 되었습니다. 당시 웹 어플리케이션을 개발하며 SQL 쿼리문을 이용한 데이터 처리에만 익숙했던 저에게 빅데이터를 다루는 분산 처리 기술이 생소했지만 깊게 파고들수록 흥미를 느꼈고, 관련 외부교육을 수강하며 빅데이터 시스템 구축을 위한 역량을 갖추고자 노력해왔습니다. 교육 과정을 통해 NoSQL 과 MongoDB, Hadoop, spark를 이용하여 Hadoop 에코시스템 구축하는 경험을 하고 싶습니다.

두번째로, 딥러닝에 입문하기 위해 Kaggle에서 진행 중인 딥러닝 공모전에 도전하였습니다. '영화 리뷰 분석'을 주제로, 영화의 리뷰가 추천인지 아닌지를 예측하기 위해 딥러닝 기법인 Word2Vec을 활용하여 데이터를 전처리하는 과정에 있습니다.
그 과정에서 자연스럽게 자연어 처리에 관심을 갖게 되었고, 챗봇을 만드는 세미나에 참여하면서 Facebook API와 Heroku서버를 활용한 챗봇 서비스에서 시작하여, 카카오톡 API와 Ngrok서버를 활용한 에코봇, 번역봇, RNN알고리즘으로 학습시킨 챗봇 서비스를 개발하였습니다. 이를 통해, 챗봇 서비스의 가능성과 한계를 알 수 있었고, 자연어 처리에 딥러닝을 접목하여 성능을 높이고 싶었습니다. 교육 과정을 통해 NLTK와 자연어처리를 적용한 챗봇을 구현해보고 싶습니다.

현재, 딥러닝의 가장 기본이 되는 데이터 전처리 작업을 위해 웹 스크레이핑의 실습 코드를 한 줄씩 읽고 직접 구현하며 JSON, XML, CSV 등 다양한 형식의 데이터를 가공하는 공부 중입니다. 최근 구현한 크롤링 애플리케이션을 통해, 네이버 금융 사이트의 약 1000일 정도의 데이터를 학습 데이터와 테스트 데이터로 분리하고 알맞은 딥러닝 알고리즘을 적용하여 정답률을 높이는 목표를 가지고 학습 중입니다. 과정의 초반에 배우게 될 Python과정을 통해 데이터 전처리 작업에 빠르게 익숙해지는데 도움이 될 것이라 생각합니다.
