```python
file2 = open('hello.txt', 'r')

# open은 text나 bytes(8bit)랑 관련된 애.
# 인코딩된건 사람을 위한 문자. 인코딩 : bytes를 특정방식으로 변환.
# bytes는 컴터를 위한 문자. 0과1. 바이트가 기본으로 저장되고 통신되는 데 쓰임.
# Python은 unicode가 기본. 사람중심으로 바뀌게됨.
# content는 byte로 반환함 => encoding방식에 따라 자유롭게 다 바꿀수있음.(utf-8, cp889, euc-kr 등등)

next(file2)
```
